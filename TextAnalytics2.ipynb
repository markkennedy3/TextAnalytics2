### Tokenization & Cleaning

#### Define Functions which will Tokenize and Clean the Data

We first need to tokenize the data. To do this we split the text in each document into a list of words. These words are called tokens and they will serve as the data input for our . 

Our data is currently unstructured. After we tokenize the data we then clean it by converting uppercase characters to lowercase and by removing special characters, stopwords and numbers. This will ensure that each word in the bag of words will be treated equally. 

Pre-Processing steps which we will apply to our data include:

- Extract tokens

- Remove stopwords

- Decapitalise words

- Remove salutation

- Remove numbers

- Lemmatization

Note: We removed special characters from the text when we imported the data.

We must create functions which will carry out each of these steps. We decided it was best to create individual functions, as this would allow us to check if they function properly (unit testing).

We'll start by defining a 'convert_tokens' function which we will use to split the content strings into seperate tokens for each word.

```python
# First: Tokenization
def convert_tokens(rawtext):
    
    ## This line compress the words with dashes or apostrophes in 
    ## them in order to keep their meaning. (eg e-mail = email)
    rawtext = rawtext.replace("'","").replace("-","")
    
    pattern = r'\w+'
    tokenizer = RegexpTokenizer(pattern)
    token_words = tokenizer.tokenize(rawtext)
    
    return token_words
```

The 'decapitalize' function shown below takes a list of tokens and returns the lower case form of each token.

```python
# Second: Decapitalization
def decapitalize(token_words):
    
    lowercase_words = [word.lower() for word in token_words]
    
    return lowercase_words
```

The function below is used to remove salutations from the text. 

We define salutaions in this cell as titles a person may have (e.g. Mr, Mrs)

```python
## Fourth : Remove salutation
def remove_salutation(token_words):
    
    salutation = ['mr','mrs','mss','dr','phd','prof','rev', 'professor','sir']
    salutation_removed = ([word for word in token_words if word.lower() not in salutation])
    
    return salutation_removed
```

The 'remove_numbers' function defined below is used to remove tokens that consist soley of numeric characters, as well as tokens with both numbers and letters in them (e.g. 3G).

```python
## Fifth: Remove Numbers
def remove_numbers(token_words):
    
    ## Not only does this remove numbers but it also removes words with numbers in them. eg 100th 
    numbers_removed = [word.lower() for word in token_words if not any(char.isdigit() for char in word)]
    
    return numbers_removed
```

The final function we will define is a 'Lemmatization' function.

"Lemmatization in is the process of grouping together the inflected forms of a word so they can be analysed as a single item" e.g. Cars -> Car 

```python
## define transfer tag function:
def transfer_tag(pos_tag):
    if pos_tag.lower().startswith('j'):
        return 'a'
    elif pos_tag.lower().startswith('v'):
        return 'v'
    elif pos_tag.lower().startswith('n'):
        return 'n'
    elif pos_tag.lower().startswith('r'):
        return 'r'
    else:
        # Default POS for lemmatization is noun.
        return 'n'
## Sixth: Lemmatization
def Lemmatization(token_words):
    
    lemmatizer = nltk.stem.WordNetLemmatizer()
    
    lemmatized_words = []
    for word, tag in nltk.pos_tag(token_words):
        wordnet_tag = transfer_tag(tag)
        lemmatized_words.extend([lemmatizer.lemmatize(word, wordnet_tag)])
        
    return lemmatized_words
```

### Top 10 Words For Each Category 

We thought it best to create a function to make the generation of these ngrams more efficient. This function creates a list of the top-10 ngrams using the corpus when given a set of tokens. 

```python
def ngram_generator(text, n):
    H1_ngram = ngrams(text, n)

    # ## Get freqs
    H1_ngram_frequencies = nltk.FreqDist(H1_ngram)

    ## to dataframe
    dict_items =list(dict(H1_ngram_frequencies).items())
    df_ngramFreq = pd.DataFrame(dict_items,columns=['ngram','freq']).sort_values(by='freq', ascending=False)
    df_ngramFreq.set_index('ngram', inplace=True)
    print("*** n-grams (n = " + str(n) +") sorted by the highest Frequency ***")
    print(df_ngramFreq.head(10))
```

### Clean Representation of Documents to Numerical Representation Vectors

Term frequency (TF) denotes the frequency of a word in each document in the corpus. 
Each document in the corpus has it's own tf and this will increase as the number of occurences of that word within the document increases. 

Inverse Document Frequency (IDF) is used to calculate the weight of the rare words across all documents in the corpus.
The words that occur rarely in the corpus have a high IDF score.

Once we combine TF and IDF we get the TF-IDF score for a word in a document in the corpus. It is the product of TF and IDF. It is defined as follows:

$$ TFIDF(term) = TF * IDF $$

$$ TF(term, document) = \frac{frequency(term)}{length(document)} $$

$$ IDF(term) = \frac{number\;of\;documents\;containing\;term}{number\;of\;documents\;in\;corpus} $$

```python
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 3), stop_words='english')

list_tokens = []
for index, row in raw_trainset.iterrows():
    list_tokens.append(' '.join(row.tokens))

train_features = tfidf.fit_transform(list_tokens).toarray()
labels = raw_trainset.category_id.values
```

Overfitting is when a model is fitted too closely to the training data and so captures the detail and noise of the data. Overfitting is the result of an overly complex model with too many parameters. Overfitting can also be referred to as high variance as it prevents supervised learning algorithms from generalising beyond their training set.

The main problem with overfitting is that rather than being able to predict correctly classified instances, it just relies on its memory from the training set leading to inaccurate results. This impacts the models ability to generalise unseen data and therefore effects its performance.

There are a number of ways to prevent overfitting from occurring. Firstly, you could take a larger training set as this will be more representative of the data as a whole which in turn will be more accurate when generalising future sets of data. Next, you could remove some features from the data set which are causing a lot of noise as this noise leads to a poor model. Cross validation could also be used when training your data as it divides data into k subsets and while each k subset is used as the test set, the other k-1 subsets are put together to form a training set. ### Confusion matrix 

These 2 functions will allow us to make confusion matrices for our models.

```python
def cm2df(cm, labels):
    df = pd.DataFrame()
    # rows
    for i, row_label in enumerate(labels):
        rowdata={}
        # columns
        for j, col_label in enumerate(labels): 
            rowdata[col_label]=cm[i,j]
        df = df.append(pd.DataFrame.from_dict({row_label:rowdata}, orient='index'))
    return df[labels]
```

```python
def plot_confusion_matrix(cm, target_names, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.figure(figsize = (15,10))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(target_names))
    plt.xticks(tick_marks, target_names, rotation=45)
    plt.yticks(tick_marks, target_names)
    plt.tight_layout()
    width, height = cm.shape
    for x in range(width):
        for y in range(height):
            plt.annotate(str(cm[x][y]), xy=(y, x), 
                        horizontalalignment='center',
                        verticalalignment='center')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

### Ridge Classifier

Ridge Regression is a regression method for analyzing multiple regression data that suffer from multicollinearity or when the number of predictor variables in a set exceeds the number of observations. 

When this occurs, least squares estimates are unbiased, but their variances are large so they may be far from
the true value. 

By adding a degree of bias to the regression estimates, ridge regression reduces the standard errors.
It is hoped that the net effect will be to give estimates that are more reliable.
